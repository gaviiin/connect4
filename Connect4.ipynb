```python
'''
Gavin Zhou
6/4/2022

This is a connect 4 game. You can change the size of the board with the height and width variables,
and change the connection length with the winning length variables. Have fun!
'''

from tkinter import *
from tkinter import font

class Connect4(Frame):
    def __init__(self, master = None):
        Frame.__init__(self)
        self.configure(width = 500, height = 100)
        police = font.Font(self, size = 20, family = 'Arial')
        self.text = Label(self, text = "Yellow turn", font = police)
        self.text.grid(sticky = NSEW, pady = 20)

class Point(object):
    def __init__(self, x, y, can, color="white", bg="red"):
        self.can = can
        self.x = x
        self.y = y
        self.color = color
        self.win_length = 4
        self.turn = 1
        self.r = 30
        self.point = self.can.create_oval(self.x + 10, self.y + 10, self.x + 61, self.y + 61, fill = color, outline = "light blue")
        
    def changeColor(self, color):
        self.can.itemconfigure(self.point, fill = color)
        self.color = color

class Grid(Canvas):
    def __init__(self, master = None):
        Canvas.__init__(self)
        self.height = 12
        self.width = 12
        self.cell_size = 65
        self.configure(width = self.width * self.cell_size, height = self.height * self.cell_size, bg = "light blue")
        self.winlen = 3

        self.player = 1
        self.color = "yellow"
        self.p = []
        self.perm = True
        for i in range(0, self.height * self.cell_size - 1, self.cell_size):
            list_range = []
            for j in range(0, self.width * self.cell_size - 1, self.cell_size):
                list_range.append(Point(j, i ,self))
                
            self.p.append(list_range)
        
        self.bind("<Button-1>", self.detectColumn)

    def detectColumn(self, event):
        if self.perm:
            col = int(event.x / 65)
            line = 0
            while line < len(self.p):            
                if self.p[0][col].color == "red" or self.p[0][0].color == "yellow":
                    break
                
                if self.p[line][col].color == "red" or self.p[line][col].color == "yellow":
                    self.p[line - 1][col].changeColor(self.color)
                    break
                
                elif line == len(self.p) - 1:
                    self.p[line][col].changeColor(self.color)
                    break
   
                if self.p[line][col].color != "red" and self.p[line][col].color != "yellow":
                    line += 1
            
            if self.player == 1:
                self.player = 2
                connect4.text.config(text="Red turn")
                self.color = "red"

            elif self.player == 2:
                self.player = 1
                connect4.text.config(text="Yellow turn")
                self.color = "yellow"
            self.vertical()
            self.horizontal()
            self.diagonal1()
            self.diagonal2()
      
    def vertical(self):
        i = 0
        print("vert")
        while(i < len(self.p) - (self.winlen)):
            j = 0
            while(j < len(self.p[i])):
                red_counter = 0
                yel_counter = 0
                k = 0
                while (k <= self.winlen):
                    if (self.p[i + k][j].color == "red"):
                        red_counter += 1
                        yel_counter = 0
                        if (red_counter == self.winlen):
                            connect4.text.config(text = "Red Player Wins!")
                            self.perm = False
                            break
                    if (self.p[i + k][j].color == "yellow"):
                        yel_counter += 1
                        red_counter = 0
                        if (yel_counter == self.winlen):
                            connect4.text.config(text = "Yellow Player Wins!")
                            self.perm = False
                            break
                    else:
                        red_counter = 0
                        yel_counter = 0
                    k += 1
                j += 1
            i += 1
            
    def horizontal(self):
        i = 0
        print("hor")
        while(i < len(self.p)):
            j = 0
            while(j < len(self.p[i]) - (self.winlen)):
                red_counter = 0
                yel_counter = 0
                k = 0
                while (k <= self.winlen):
                    if (self.p[i][j + k].color == "red"):
                        red_counter += 1
                        yel_counter = 0
                        if (red_counter == self.winlen):
                            connect4.text.config(text = "Red Player Wins!")
                            self.perm = False
                            break
                    if (self.p[i][j + k].color == "yellow"):
                        yel_counter += 1
                        red_counter = 0
                        if (yel_counter == self.winlen):
                            connect4.text.config(text = "Yellow Player Wins!")
                            self.perm = False
                            break
                    else:
                        red_counter = 0
                        yel_counter = 0

                    k += 1
                j += 1
            i += 1

    def diagonal1(self):
        i = 0
        print ("diag1")
        while(i < len(self.p) - (self.winlen)):
            j = 0
            while(j < len(self.p[i]) - (self.winlen)):
                red_counter = 0
                yel_counter = 0
                k = 0
                while (k <= self.winlen):
                    if (self.p[i + k][j + k].color == "red"):
                        red_counter += 1
                        yel_counter = 0
                        if (red_counter == self.winlen):
                            connect4.text.config(text = "Red Player Wins!")
                            self.perm = False
                            break
                    if (self.p[i + k][j + k].color == "yellow"):
                        yel_counter += 1
                        red_counter = 0
                        if (yel_counter == self.winlen):
                            connect4.text.config(text = "Yellow Player Wins!")
                            self.perm = False
                            break
                    else:
                        red_counter = 0
                        yel_counter = 0

                    k += 1
                j += 1
            i += 1
            
    def diagonal2(self):
        i = 0
        print ("diag2")
        while(i < len(self.p) - (self.winlen)):
            j = len(self.p[1]) - 1
            while(j > self.winlen):
                red_counter = 0
                yel_counter = 0
                k = 0
                while (k <= self.winlen):
                    print(i + k, j + k)
                    if (self.p[i][j].color == self.p[i + k][j - k].color == "red"):
                        red_counter += 1
                        yel_counter = 0
                        if (red_counter == self.winlen):
                            connect4.text.config(text = "Red Player Wins!")
                            self.perm = False
                            break
                    if (self.p[i][j].color == self.p[i + k][j - k].color == "yellow"):
                        yel_counter += 1
                        red_counter = 0
                        if (yel_counter == self.winlen):
                            connect4.text.config(text = "Yellow Player Wins!")
                            self.perm = False
                            break
                    else:
                        red_counter = 0
                        yel_counter = 0

                    k += 1
                j -= 1
            i += 1

root = Tk()
root.state('zoomed')
root.title("Gavin Connect 4")

connect4 = Connect4(root)
connect4.grid(row = 0, column = 0)


t = Grid(root)
t.grid(row = 1, column = 0)

def execute():
    global connect4
    connect4.text.config(text = "")
    
    connect4 = Connect4(root)
    connect4.grid(row = 0, column = 0)

    t = Grid(root)
    t.grid(row = 1, column = 0)

Button(root, text = "Restart", command = execute).grid(row = 2, column = 0, pady = 30)

root.mainloop()

```
